```python
data_num= tiktok_data.copy()
data_num.isnull().sum()
```

```python
# Columns I need For EDA and sentiment dashboard.

# authorMeta/fans:  Total number of followers the user has.
# authorMeta/heart: Total number of likes the user has received across all their videos.
# authorMeta/name: Username of the user .
# authorMeta/signature: User’s bio or profile description.
# authorMeta/verified: Whether the user is verified (True or False).
# authorMeta/video: Total number of videos posted by the user.

# commentCount: Number of comments on the video.
# diggCount: Number of likes (or "diggs") the video received.
# downloaded:Indicates whether the video was downloaded (True or False).
# playCount: Total number of times the video was played.
# shareCount:Number of times the video was shared.

# hashtags/0/name: Name of the first hashtag .
# hashtags/0/title: Title of the first hashtag campaign (if any).
# hashtags/1/name	Name of the second hashtag.
# hashtags/1/title:Title of the second hashtag campaign.

# mentions/0	First mentioned user (if any).

# musicMeta/musicOriginal	Whether the music is original (True or False).

# searchHashtag/name	The main hashtag used in the video.
# searchHashtag/views	Total views for this hashtag across TikTok.

# text	Caption or description of the video.
# videoMeta/duration	Length of the video in seconds.
# videoMeta/height	Video resolution height in pixels.
# videoMeta/width	Video resolution width in pixels.
```

```python
data_num.drop(["authorMeta/avatar", "authorMeta/digg", "authorMeta/following", "authorMeta/id", "authorMeta/nickName","effectStickers/0/ID","effectStickers/0/name",
"effectStickers/0/stickerStats/useCount",  "effectStickers/1/ID", "effectStickers/1/name", "effectStickers/1/stickerStats/useCount",  "hashtags/0/cover", 
"hashtags/1/cover", "id", "mediaUrls/0", "musicMeta/playUrl", "createTime", "createTimeISO","mentions/1","mentions/2","mentions/3","mentions/4",
"mentions/5","musicMeta/musicAlbum"	,"musicMeta/musicAuthor" , "musicMeta/musicName","hashtags/0/id","hashtags/1/id","webVideoUrl" ],axis=1,inplace=True)
data_num.isnull().sum()
```

```python

data_num["mention"] = data_num["mentions/0"].notna().astype(int)
data_num["bio"]=data_num["authorMeta/signature"].notna().astype(int)
def classify_orientation(width,height):
    if width > height:  # Landscape (wider)
        return 'Landscape'
    else :  # Portrait (taller)
        return 'Portrait'

data_num['orientation'] = data_num.apply(lambda row: classify_orientation(row['videoMeta/width'], row['videoMeta/height']), axis=1)
data_num['orientation'].unique()
```

```python
data_num.to_excel('filtered_data.xlsx', sheet_name='MySheet', index=False)
data_num.to_csv('filtered_data_csv.csv', index=False)
final=data_num.copy()
def format_num(num):
    if num >=1000000000:
        return f"{num/1000000000:.2f}B"
    elif num >=1000000:
        return f"{num/1000000:.2f}M"
    elif num >=1000:
        return f"{num/1000:.2f}K"
    else:
        return str (num)
final['Formatted_fans']=final['authorMeta/fans'].apply(format_num)
final['Formatted_heart']=final['authorMeta/heart'].apply(format_num)
final['Formatted_video']=final['authorMeta/video'].apply(format_num)
final['Formatted_diggCount']=final['diggCount'].apply(format_num)
final['Formatted_commnet']=final['commentCount'].apply(format_num)
final['Formatted_playCount']=final['playCount'].apply(format_num)
final['Formatted_shareCount']=final['shareCount'].apply(format_num)
final.isnull().sum()

```

```python
final.drop(['authorMeta/fans','authorMeta/heart','authorMeta/video','diggCount','commentCount','playCount',
           'shareCount'],axis=1,inplace=True)
final.to_excel('final_data.xlsx', sheet_name='MySheet', index=False)

```
